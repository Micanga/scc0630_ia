# APRIORI ASSOCIATION RULE IMPLEMENTATION

Hello everyone. Today we will implement the Association Rule Apriori (AR-Apriori) Algorithm.

The first step is to import useful packages for this implementation.

For this task, we will use the <i>apyori</i> Python Package and its method <i>apriori</i>, mainly.

```python
# Importing the packages
from apyori import apriori 
```

This method implements three major components of the Apriori algorithm:
- Support Calculation;
- Confidence Calculation, and;
- Lift Calculation;

Furthermore, it also provides parameters to adjust our rules extraction - as the minimum/max confidence, support, lift, etc.

## 1. Preprocesing the Dataset

After the import, we will preprocess the available data set to apply it to the given method.

```python
# 1. Preprossesing (file to list of lists)
dotto = list()
with open('dotto.data') as  dotto_file: 
    for line in dotto_file:
        # a. transforming the file line in a formated
        # transaction (list of words)
        transaction = [word.replace('\n','') for word in line.split(' ')]
        
        # b. appending the formated transaction
        dotto.append(transaction)
        
        # c. printing it
        print '( Transaction',len(dotto),') - Number of items:',len(transaction),'- Items:',
        for item in transaction:
            print item,
        print
```

# 2. Association Rules Extration

With the preprocessed data, we start to extract the association rules applying the aprori technique.

To verify which values to support, confidence and lift to use that possibly give a better output, we will run the <i>apriori method</i> for different parameter values (which we consider important).

The number of association created will be analysed to choose the best parameter configuration.

```python
# 2. Extracting the rules
association_rules = dict()
for sup in [0.005,0.01,0.05,0.1]:
    for conf in [0.7,0.8,0.9]:
        for lift in [2,3,4,5,6,7]:
            print 'AR_Apriori(sup=',sup,',conf=',conf,',lift=',lift,')'
            association_rules[str(sup)+'_'+str(conf)+'_'+str(lift)] = list(apriori(dotto, min_support=sup, min_confidence=conf, min_lift=lift))
            print '- Number of association rules found:', len(association_rules[str(sup)+'_'+str(conf)+'_'+str(lift)])
            print '========================================='
```

Analysing the output, we will use the parameters with configuration equals to <i>(support = 0.01, confidence = 0.7, lift = 3)</i> to identify and split trivial knowledge from new knowledge.

# 3. Rules Analysis

With the defined parameters, we will analyze the rules generated, trying to identify and split the trivial knowledge (previous knowledge) from new knowledge (obtained after AR-Apriori algorithm).

The follow script print on the screen the created association rules.

```python
# 3. Showing the result
sup, conf, lift = 0.01, 0.7, 3
for item in association_rules[str(sup)+'_'+str(conf)+'_'+str(lift)]:
    # a. getting the pair base-add of items
    pair = item[2] 
    base = [it for it in pair[0][0]]
    add = [it for it in pair[0][1]]

    # b. printing the rule
    print("Rule: " + str(base) + " -> " + str(add))
    
    # c. printing the rule components
    print("Support: " + str(item[1]))
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
